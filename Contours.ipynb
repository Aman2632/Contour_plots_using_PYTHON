```python
# Q_6 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a. ROSENBROCK 

rosenbrock = lambda x,y: (1-x)**2+100*(y-x**2)**2

n = 100       # ploting for 1000 points  along the x-axis
m = 100      # ploting for 1000 points   along the x-axis
a=-0.5; b=2. # extreme in the x-axis
c=-1.5; d=4. # extreme in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Z = rosenbrock(X,Y)

plt.contour(X,Y,Z,np.logspace(-0.5,3.5,20,base=10))
#plt.contourf(X,Y,Z)
plt.title("Rosenbrock contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# b. Rastrigin:



a= (-5.12); b= 5.12 # Upper and lower bounds of x
d= 2                #Since we are plotting 2-D contours

X,Y = np.meshgrid(np.linspace(a,b,100),np.linspace(a,b,100))

Z= 10*d + X*X - 10*np.cos(2*np.pi*X) + Y*Y - 10*np.cos(2*np.pi*Y)


plt.contour(X,Y,Z)
#plt.contourf(X,D,Z)
plt.title("Rastrigin contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# c. Booth's Contour


n = 100       # ploting for 1000 points  along the x-axis
m = 100      # ploting for 1000 points   along the x-axis
a=-10; b=10. # extreme in the x-axis
c=-10; d=10. # extreme in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Booth = (X + 2*Y -7)**2 + (2*X + Y -5)**2

plt.contour(X,Y,Booth,np.logspace(-10,10,100,base=10))
#plt.contourf(X,Y,Z)
plt.title("Booth contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# d. Beale's Contour


n = 100       # ploting for 1000 points  along the x-axis
m = 100      # ploting for 1000 points   along the x-axis
a=(-4.5); b=4.5 # extreme in the x-axis
c=-4.5; d=4.5 # extreme in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Beale = (1.5 - X + X*Y)**2 + (2.25 - X + X*Y*Y)**2 + (2.625 - X + X*(Y**3))**2

plt.contour(X,Y,Beale,np.logspace(-4.5,4.5,50,base=10))
#plt.contourf(X,Y,Z)
plt.title("Beale contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# e. McCormick's Contour


n = 100       # ploting for 1000 points  along the x-axis
m = 50     # ploting for 1000 points   along the x-axis
a=(-1.5); b=4 # extreme in the x-axis
c=-3; d=4 # extreme in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

McCormick = np.sin(X + Y) + (X-Y)**2 - 1.5*X + 2.5*Y +1
plt.contour(X,Y,McCormick,np.logspace(-1.5,4,90,base=10))
#plt.contourf(X,Y,Z)
plt.title("McCormick contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# f. Eggholdar's Contour


n = 100         # ploting for 1000 points  along the x-axis
m = 100         # ploting for 1000 points   along the x-axis
a=(-512); b=512 # extreme in the x-axis
c=(-512); d=512 # extreme in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Eggholdar = -(Y*Y+47)*np.sin(np.sqrt(np.abs(Y+(X/2)+47))) - X*np.sin(np.sqrt(np.abs(X - (Y+47))))
plt.contour(X,Y,Eggholdar)
#plt.contourf(X,Y,Z)
plt.title("Eggholdar contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# g. Drop-Wave's Contour


n = 100         # ploting for 1000 points  along the x-axis
m = 100         # ploting for 1000 points   along the x-axis
a=(-5.12); b=5.12 # extreme in the x-axis
c=(-5.12); d=5.12 # extreme in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m));

Drop_Wave = (-1*((1+ np.cos(12*np.sqrt(X*X + Y*Y)))/(0.5*(X*X + Y*Y) +2)))
plt.contour(X,Y,Drop_Wave)
#plt.contourf(X,Y,Z)
plt.title("Drop-Wave contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# h. Three_hump_camel's Contour Plot

n = 100         # ploting for 1000 points  along the x-axis
m = 100       # ploting for 1000 points   along the x-axis
a=(-5); b=5 # extreme in the x-axis
c=(-5); d=5 # extreme in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m));

Three_hump_camel = 2*X**2 - 1.05*X**4 + (X**6)/6 + X*Y + Y**2
plt.contour(X,Y,Three_hump_camel,np.logspace(-5,5,90,base=10))
#plt.contourf(X,Y,Z)
plt.title("Three_hump_camel contour")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
